**Naive Bayes**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/kb22/Heart-Disease-Prediction/master/dataset.csv')
dataset.head(5)
```

```python
X = dataset.iloc[:,:14].values
y = dataset['target'].values
dataset.head(5)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test) 
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
from sklearn.metrics import accuracy_score 
print ("Accuracy : ", accuracy_score(y_test, y_pred))
cm
```

**Kesimpulan** : berikut prediksi "Target" menggunakan naive bayes

```python
df = pd.DataFrame({'Real Values':y_test, 'Predicted Values':y_pred})
df
```

**KNN**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/kb22/Heart-Disease-Prediction/master/dataset.csv")
df.head()
```

Melihat ukuran dataset

```python
df.shape
```

Membagi data ke x dan y

```python
# X tanpa label
X = df.drop(columns=["target"])
X.head()
```

```python
#y labelnya saja
y = df["target"].values
y[0:303]
```

Membagi data ke data test dan data training.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0, stratify=y)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_train[:5]
```

Membangun knn classifier

```python
knn = KNeighborsClassifier(n_neighbors = 5,metric = 'minkowski', p = 2)
knn.fit(X_train,y_train)
```

```python
KNeighborsClassifier()
```

```python
knn.predict(X_test)[0:14]
```

```python
#check accuracy of our model on the test data
knn.score(X_test, y_test)
```

Berikut adalah hasil akurasi:

```python
from sklearn.metrics import *
import numpy as np

#Check Accuracy precision, recall, f1-score
print( classification_report(y_test, knn.predict(X_test)) )
#Another way to get the models accuracy on the test data
print(F'Accuracy:',accuracy_score(y_test, knn.predict(X_test)))
print(F'Precision:', precision_score(y_test, knn.predict(X_test)))
print(F'Recall:', recall_score(y_test, knn.predict(X_test)))
print(F'F1 Score:', f1_score(y_test, knn.predict(X_test)))

#Check Roc Auc Score
print( F'Roc Auc Score:',roc_auc_score(y_test, knn.predict(X_test)) )

print( F'Balanced Accuracy Score:',balanced_accuracy_score(y_test, knn.predict(X_test)) ) 
print( F'Confusion Matrix:',confusion_matrix(y_test, knn.predict(X_test)) )
print()#Print a new line
```
